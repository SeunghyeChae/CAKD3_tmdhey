```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

import re
def dong(x):
    import re

    p = re.compile('[가-힣]\S*[읍면동가]')

    r = p.findall(x)
    
    return ''.join(r)


```

```python
#
u1 = ['공릉동', '상계동', '월계동', '중계동', '하계동']
시도명 = '서울특별시'
시군구명 = '동대문구' 
제공기관명 = f'{시도명} {시군구명}'
# u1 = '검색하실 읍/면/동'
```

```python
df= pd.read_csv('전처리완료dataset/human/서울특별시.csv')

# df = df.rename(columns=df.iloc[0])
# df.drop('Unnamed: 0',axis=1,inplace=True)


df
```

```python
detail_1 = df[df['시군구명']==시군구명]
detail_1['읍/면/동'] = detail_1['읍면동'].apply(lambda x : dong(x))
```

```python

시도명= 제공기관명.split(" ")[0]
시군구명= 제공기관명.split(" ")[1]
detail_1['시도명']= 시도명
detail_1['시군구명']= 시군구명
get_col = ['시도명','시군구명','읍/면/동']
detail_1 = detail_1[get_col]
detail_1['어린이보호구역총계'] = 1
grouped = detail_1.groupby('읍/면/동')[['어린이보호구역총계']].sum()
result = grouped.reset_index()
result['시도명'] = '서울특별시'
result['시군구명'] = '노원구'
result = result[['시도명','시군구명','읍/면/동','어린이보호구역총계']]
result
```
