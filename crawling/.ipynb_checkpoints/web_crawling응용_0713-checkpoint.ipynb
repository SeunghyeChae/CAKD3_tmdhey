DOM(Document ObjectModel)의 정의

- HTML, XML 문서의 프로그래밍 인터페이스: 구조화된 표현 및 프로그래밍 언어가 DOM 구조에 접근할 수 있는 방법을 제공
- 트리 구조로 형성되어 있음: 부모노드(위쪽), 자식노드(아래쪽)
- HTML 에서 노드는 head,body,h1,script 등의 태그뿐만 아니라 태그 내 텍스트나 속성 모두 노드에 속함
- BeautifulSoup 모듈의 함수를 활용하여 노드를 기준으로 원하는 데이터 추출 

```python
html= '''
<html>
<head>
    <title>crawler</title>
</head>
<body>
    <p class="a" align="center"> text1 </p>
    <p class="b" align="center"> text2 </p>
    <p class="c" align="center"> text3 </p>
    <div>
        <img src="/source" width="300" height="200">
    </div>
</body>
</html>
'''

from bs4 import BeautifulSoup

bs= BeautifulSoup(html,'html.parser')
contents= bs.find('body')

for child in contents.children:
    print(child)
```

```python
# body의 자손은 p ,div, img
for d in contents.descendants:
    print(d)
```

```python
img_tag= contents.find('img')
print(img_tag)
print(img_tag.parent)
```

```python
contents=bs.find('body')
print(img_tag.find_parent('body'),'\n')
print(img_tag.find_parent('div'))

#특정 부모 찾을 때 find_
```

```python
p_tag= bs.find('p',class_='b')   #예약어와의 충돌을 피하기 위해 class_ 
print(p_tag)
```

```python
from urllib import request as req
from bs4 import BeautifulSoup
res = req.urlopen("https://naver.com")
bs = BeautifulSoup(res,'html.parser')
print(bs.find('a'),'\n')
print(bs.find(class_='link_newsstand'),'\n')

eles= bs.find_all('a',{'class':['link_newsstand','btn_sort','btn_sort.sort_on']}) #클래스가 여러개일 경우
for e in eles:
    print(e.text)
```

```python
hlist= bs.findAll({'h1','h2','h3','h4','h5','h6'},limit=3)
for h in hlist:
    print(h, '\n')
```

```python
# 정규표현식과 bs4
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html= urlopen('https://pythonscraping.com/pages/page3.html')
bs= BeautifulSoup(html,'html.parser')
images=bs.find_all('img',{'src':re.compile('\.\.\/img/gifts/img.*\.jpg')}) 
# . 은 아무문자나 상관x , *은 0개이상/   정규표현식 아닌 기호 그대로 쓸 때 앞에 \ 붙이기 

for image in images:
    print(image['src'])
```

```python
# Q. 구글 플레이에서 인기 영화 제목 30개 출력( requests+ bs4사용)
# header.지정 
import requests
from bs4 import BeautifulSoup

headers= {
    'Accept-Language':'ko-KR.ko'
}

url='https://play.google.com/store/movies/top'
# url= 'https://play.google.com/store/movies/top?hl=ml&gl=KR'


res=requests.get(url, headers=headers)
res.raise_for_status()
bs=BeautifulSoup(res.text, 'html.parser')

movies= bs.find_all('div',attrs={'class':'WsMG1c nnK0zc'})

print(len(movies))
for m in movies:
    print(m.text)
```

```python
# 구글 플레이에서 인기 영화제목 200개 출력(selenium_bs4)
from selenium import webdriver

driver= webdriver.Chrome('C:/♥workspace/tool/chromedriver.exe')
driver.maximize_window()

url= 'https://play.google.com/store/movies/top'
driver.get(url)

#한페이지만큼내리기 (1080 위치로 스크롤 내리기)
driver.execute_script('window.scrollTo(0,1080)') 
#화면 가장 아래로 스크롤 내리기 (자바스크립트)
# driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')

import time 
interval= 2 #2초에 한 번씩 스크롤 내리기

# 현재 문서 높이를 가져와서 저장 
prev_height= driver.execute_script('return document.body.scrollHeight')

# 반복 수행 
while 1: 
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    # 페이지 로딩 대기
    time.sleep(interval)
    curr_height = driver.execute_script('return document.body.scrollHeight')
    # 비교
    if curr_height == prev_height:
        break
    else:
        prev_height = curr_height
        
        
bs = BeautifulSoup(driver.page_source, 'html.parser')
titles = bs.select('.WsMG1c.nnK0zc')
for title in titles:
    print(title.text)
```

```python
# headless chrom

from selenium import webdriver

options= webdriver.ChromOptions()
options.headless= True
options.add_argument('window-size=1920x1080')


driver= webdriver.Chrome('C:/♥workspace/tool/chromedriver.exe', options=options)
driver.maximize_window()

url= 'https://play.google.com/store/movies/top'
driver.get(url)

#한페이지만큼내리기 (1080 위치로 스크롤 내리기)
driver.execute_script('window.scrollTo(0,1080)') 
#화면 가장 아래로 스크롤 내리기 (자바스크립트)
# driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')

import time 
interval= 2 #2초에 한 번씩 스크롤 내리기

# 현재 문서 높이를 가져와서 저장 
prev_height= driver.execute_script('return document.body.scrollHeight')

# 반복 수행 
while 1: 
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    # 페이지 로딩 대기 
    time.sleep(interval)
    curr_height= driver.execute_script('return document.body.scrollHeight')
    #비교
    if curr_height == prev_height:
        break 
    prev_height = curr_height
print('스크롤 완료')

bs=BeautifulSoup(driver.page_source, 'html.parser')
movies= bs.find_all('div',attrs={'class':'.WsMG1c.nnK0zc'})

print(len(movies))
for m in movies:
    print(m.text)
```
