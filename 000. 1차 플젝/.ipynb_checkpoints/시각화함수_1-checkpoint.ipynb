```python
df1 = pd.read_pickle('dataset/purprd_제휴사/purprd_A.pkl')
df2 = pd.read_pickle('dataset/purprd_제휴사/purprd_B.pkl')
df3 = pd.read_pickle('dataset/purprd_제휴사/purprd_C.pkl')
df4 = pd.read_pickle('dataset/purprd_제휴사/purprd_D.pkl')
```

```python
import seaborn as sns
# 데이터프레임 슬라이싱 / a1 = 1월~12월 매출  , a2 = 1월~12월 매출증감
# df에는 df명 index에는 인덱스의 숫자 번호가 들어가야함
def slice_df(df,index): 
    df1 = df.iloc[[index]].iloc[:,:12] ## 수정요망
    df2 = df.iloc[[index]].iloc[:,12:25] ## 수정요망
    df3 = df.iloc[[index]].iloc[:,25:35]
    df4 = df.iloc[[index]].iloc[:,35:-1]
    index = df.index[index-1]
    
    return df1, df2 ,df3, df4 , index

# 슬라이싱된 두 데이터(매출총합, 매출증감)을 각각 시각화 
def visualization_sales_per(df1, df2, df3, df4, index):
    import matplotlib.pyplot as plt
    from matplotlib import rc
    import numpy as np
    
    plt.style.use('seaborn-dark')
    %config InlineBackend.figure_format = 'retina'
    rc('font',family='Malgun Gothic') # 한글 폰트 설정
    plt.rcParams['axes.unicode_minus'] = False #
    
    
    def bar_color(df,color1,color2):
        return np.where(df.values>0,color1,color2).T
    
    fig, ax = plt.subplots(ncols=4,figsize=(25, 6))
    
    sns.barplot(data=df1,ax=ax[0])
    sns.barplot(data=df2,ax=ax[1])
    sns.barplot(data=df3,ax=ax[2])
    sns.barplot(data=df4,ax=ax[3])
    
    # 타이틀
    ax[0].set_title(f'{index} 의 2014년 매출',fontsize=12,fontweight='bold')
    ax[1].set_title(f'{index} 의 2015년 매출',fontsize=12,fontweight='bold')
    ax[2].set_title(f"{index} 의 2014년 변화율",fontsize=12,fontweight='bold')
    ax[3].set_title(f"{index} 의 2015년 변화율",fontsize=12,fontweight='bold')
    

# 두 함수 종합 (데이터프레임, 보고자하는인덱스 넣으면 데이터프레임의 해당 인덱스(한개)에 대한 그래프 반환)  
def df_visualization(df,index):
    df1, df2, df3, df4,index = slice_df(df,index)
    
    visualization_sales_per(df1,df2,df3,df4,index)
    

# 얘만 쓰면 다 돌아가요, x에 데이터프레임 y에 인덱스   ex)))) df_visualization(x,y)
df_visualization(a_all, 0)
```

```python
##########################################################################################################################
##########################################################################################################################
################################################ A의 대분류코드 별 판매 증감추이 ################################################
##########################################################################################################################
##########################################################################################################################

for i in range(len(a_cat)):
    df_visualization(a_cat, i)
```

```python
##########################################################################################################################
##########################################################################################################################
################################################ A의 중분류명별 판매 증감추이 ################################################
##########################################################################################################################
##########################################################################################################################

for i in range(len(a_mid)):
    df_visualization(a_mid, i)
```

```python
##########################################################################################################################
##########################################################################################################################
################################################ A멤버십 전체 판매 증감추이 ################################################
##########################################################################################################################
##########################################################################################################################

df_visualization(a_all, 0)
```
