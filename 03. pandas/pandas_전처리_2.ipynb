```python
# Age column을 범주화하여 age_class 컬럼을 생성하세요
len(tdf.age.unique())
#categorization : 범주화 필수
```

```python
tdf['age'] = tdf['age'].fillna(tdf['age'].mean().round(0)) 
#nan값을 평균값으로 대체
```

```python
def categorization_age(x):
    cat = ''      #category 지정
    if x < 10: 
        cat ="children"
    elif 10<= x <20 : 
        cat = "teen"
    elif 21<= x <40 : 
        cat ="young_adults"
    elif 40<= x <59 :  
        cat = "adults"
    else: 
        cat = "senior"
    return cat
#함수설정
```

```python
#데이터프레임 열에 함수를 적용
ag = tdf.age.apply(lambda x : categorization_age(x))
tdf['age_class'] = ag
```

```python
#함수확인
print(tdf[['age','age_class']])
tdf['age_class'].value_counts()
```

```python
# height의 값을 cm로 변환하여 새로운 컬럼 height_cm 생성 

def ftinch_to_cm(ftinch):
    ftinchlist = ftinch.split("'")
    ft = int(ftinchlist[0])
    inch = int(ftinchlist[1])
    ftcm=ft*30
    incm = inch*2.54
    total = ftcm + incm
    return total


fifa['Height_cm']=fifa.Height.apply(lambda x:ftinch_to_cm(x))
fifa.head()
```

```python
# Height 값을 cm로 변환하여 새로운 컬럼 Height_cm을 생성하세요.
f_df = fifa.copy()

def ht_ex(x):
    f, i = map(int, x.split('\''))
    return f*30.48 + i*2.54

f_df['Height_cm'] = f_df['Height'].apply(ht_ex)
f_df.head()

```
