```python
# 구단별 max웨이지 확인 / 내림차순
wage_max_club= fifa[['Club','Wage']].groupby('Club').max()
wage_max_club.sort_values('Wage',ascending=False)
```

```python
# 구단별 평균 wage (club별로 평균내기)
wage_aver_club= fifa[['Club','Wage']].groupby('Club').mean().round(2)

wage_aver_club.sort_values('Wage',ascending=False)
```

```python
# 구단별 평균 wage 추가! 

fifa1 = fifa.copy()
```

```python
a = fifa1.groupby('Club').Wage.mean().round(2)
fifa['mean_club_wage'] = fifa['Wage']
def get_mean(x):
    return a[x]
```

```python
fifa1['mean_club_wage'] = fifa1['Club'].apply(get_mean)
fifa1.head()
```

```python
# 구단별 추가 wage 추가 2
```

```python
wage_av_c = fifa[['Club','Wage']].groupby('Club').mean().round(2)
wage_av_c['avW'] = wage_av_c["Wage"]
wage_av_c.drop('Wage',axis=1,inplace=True)
```

```python
wage_av_c.sort_values('avW',ascending=False)
wage_av_c.reset_index(inplace=True)
```

```python
# 키 cm바꿔서 합치기
def hei(x):
    f, i = map(int, x.split('\''))
    return f*30.48 + i*2.54

fifa['Height_cm'] = fifa['Height'].apply(hei)
fifa.head()
```

```python
# 몸무게 kg으로 바꿔서 합치기
fifa['Weight_kg'] = (fifa.Weight_lb * 0.453592).round(1)

fifa.head()
```
