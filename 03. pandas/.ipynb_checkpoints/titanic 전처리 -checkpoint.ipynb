독립변수에 의해서 결정되는게 종속변수니까,
독립변수(전처리한것)을 종속변수(test,검증데이터)에 맞춰서 학습하는걸거에요..!
독립변수(train)에서 사람들을 쭈우욱 정리를 해두고
종속변수(test, 전처리되지 않은 데이터)에 적용이 됬을때
종속변수 사람들이 살았냐 죽었냐를 여기 머신러닝이 학습을 하는거죠

그러면 독립변수(train) 종속변수(test)변수가
얼마나 정확하게 되었냐가

머신러닝의 학습의 정도가 나타나고
그걸 100%라면 엄청난거겠죵?


저 데이터에서는 독립변수 = Survive = Y 이고 
종속변수가 다른 인자들  = X 에요!


전처리를 무작정 한다기보다는
목표에 맞는 전처리를 해줘야만
머신러닝의 학습이 좀더 목표에 맞게 학습을 할거고


변수 선정이 키 인듯


그래서 가끔은 과감하게 버리는 것도 하나의 방법이라고 하신 게 아닐까 싶어요


예측정확도를 높이는 변수를 선택하고, 데이터셋에서 null값을 적절히 처리해주는게 숙제인것 같아요


독립변수에 따라서 종속변수가 어떤지, 

(여기선 
독립변수: class, 남여, 요금, 등등의 조합과
종속변수 0:죽음, 1:생존) 의 쌍을 학습하고

테스터데이터로
테스터 독립변수를 입력해서 종속변수(0:죽음, 1:생존)를 예측해 보고,
실제 결과랑 비교해 보는 것이에요


독립변수와 종속변수의 관계, 알고리즘을 컴퓨터가 학습해줘서 검증용 데이터에 적용해주는거 개꿀이네요

```python
def categorization_age(x):
    cat = ''
    if x < 10: 
        cat ="children"
    elif 10<= x <20 : 
        cat = "teen"
    elif 21<= x <40 : 
        cat ="young_adults"
    elif 40<= x <59 :  
        cat = "adults"
    else: 
        cat = "senior"
    return cat
#함수설정
```

```python

ag = titanic.age.apply(lambda x : categorization_age(x))
titanic['age_class'] = ag
```

```python
#함수확인
print(titanic[['age','age_class']])
titanic['age_class'].value_counts()
```
