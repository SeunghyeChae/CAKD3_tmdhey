```python
import pandas as pd
import numpy as np

np.random.seed(0)
data=np.random.randint(80,100,size=(5,5))
df=pd.DataFrame(data,columns=list('abcde'))
df
```

```python
df1= df.copy()
df1
```

```python
#특정인덱스를 넣고싶을 때

#일단 칼럼을 추가
df1['index']=list('가나다라마')
df1
```

```python
#index를 인덱스 로 넣기 (원래는 초기값으로 주어짐 )

df2=df1.set_index('index')
df2
```

```python
#다시 원상태로 돌아가고싶으면  --> 다시 컬럼으로 바뀜 
df3= df2.reset_index()
df3
```

```python
# 칼럼 순서 바꾸기  (순서대로 적으면 됨 ) (대괄호 두 개)
df1=df3[['a','b','c','d','e','index']]
df1
```

```python
#  Q. 딕셔너리로 데이터 프레임을 5행5열로 작성
index= ['r0','r1','r3','r4']
column=list('abcd')

dict_data={'a':[1,2,3,4,5],'b':[4,5,6,7,8],'c':[7,8,9,10,11],'d':[12,13,14,15,16],'e':[17,18,19,20,21]}

#딕셔너리를 ㄷ이터 프레임으로 변환 인덳스를 r0r1r2로 지정 
df= pd.DataFrame(dict_data, index=['r0','r1','r2','r3','r4'])
df
```

```python
# Q. df에 r5,r6 2 개의 행을 추가하고 값을 0을 적용하여 출력 

df.loc['r5']=0
df.loc['r6']=0
df
```

```python
# Q. df에 r8 행을 추가하고 값을 0을 적용하여 출력 

new_index= ['r0','r1','r2','r3','r4','r5','r6','r7','r8']
df5=df.reindex(new_index,fill_value=0)
df5
```

```python
# Q. df5에서 r5~r8까지 삭제후 df1으로 출력 
df1= df5.drop(['r5','r6','r7','r8'],axis=0)   #df1이아니라 df5에 즉각출력= inplace=True
df1
```

```python
#컬럼 정렬 
df1_s= df1.sort_index(ascending=False)  #아무것도 없으면 올림차순/ ascending=False : 역순 
df1_s
```

```python
# 특정 컬럼 기준으로 정렬

df1_c= df1.sort_values(by='c')
df1_c
```

```python

df1_c= df1.sort_values(by='c', ascending=False) #내림차순
df1_c
```
