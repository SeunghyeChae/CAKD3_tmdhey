## 그룹 연산 

- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나누어서 관찰할 수 있으며 이런 방식으로 분할 처리하는 것을 그룹 연산이라 함.
- 그룹 연산은 데이터를 집계, 변환, 필터링하는데 효율적이며 판다스 groupby()메소드를 사용하
- 그룹 객체 만들기(분할)
- 그룹 연산 메소드 (적용-결합)

```python
# Q. 최대값- 최소값을 반환하는 사용자 함수를 정의하고 agg()함수에 인수로
# 전달하여 그룹별로 집계하여 agg_minmax이름으로 저장 후 처음 5개행을 출력

titanic= sns.load_dataset('titanic')
df= titanic.loc[:,['age','sex','class','fare','survived']]

# class를 열 기준으로 분할 #########################################################################
grouped= df.groupby(['class'])

def minmax(x):
    return x.max() - x.min()

agg_minmax = grouped.agg(minmax)
agg_minmax

```

```python
# 모든 열에 여러 함수를 매핑: group객체.agg([함수1,함수2])
# 각 열마다 다른 함수 매핑: group객체.agg({'col1':함수1, 'col2':함수2 ...})

agg_all= grouped.agg(['min','max'])
agg_all
```

```python

agg_dif= grouped.agg({'age':'mean','fare':['min','max']})
agg_dif
```

```python
# Q. 데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터 프레임으로 반환하여
# grouped_filter 이름으로 저장 후 처음 5개행을 출력하세요 

grouped_filter = grouped.filter(lambda x : len(x) >= 200)
grouped_filter['class'].
```

```python
# Q age 열의 평균이 30보다  작은 그룹만을 필터링하여 데이터프레잉으로 반환하기
# age_filter 사용 ='/'
age_filter = grouped.filter(lambda x : x['age'].mean() < 30)
age_filter
```

```python
# z-score: 평균값에서 표준편차의 몇 배 떨어져 있는지를 평가하는 수치 (평균에서부터 값이얼마나 멀리있는지)

def z_score(x):
    return (x - x.mean()) / x.std()

age_z_score = grouped.age.apply(z_score)
age_z_score.head()
```

```python
age_zscore= grouped.age.transform(z_score)
age_zscore.head()
```

```python
age_filter= grouped.apply(lambda x : x.age.mean()<30)  #filter/ apply 다른점
# 각 그룹의 age 가 30보다 작은 그룹 필터링 -> age 필터에 저장
# for문으로 조건에 맞는것만 뽑기 
for x in age_filter.index: # 모수에서 x를 하나씩 뽑는다 (관측치들을 지정해주기 위해서 index를 넣는것..) # 조건만족하는 인덱스 
    if age_filter[x]==True:
        age_filter_df= grouped.get_group(x)
        print(age_filter_df.head())
        print()

#음.. 근데 왜 이것들밖에안뽑히는거지.. - head써서 ㅋㅋ
```
