```python
# 과제
# Q. embarked, who,adult_male에 대해 Label Encoding하세요
import seaborn as sns

titanic= sns.load_dataset('titanic')
titanic.head()

#-----------------------------------------------------------
from sklearn.preprocessing import LabelEncoder 

le= LabelEncoder()
features= ['embarked','who','adult_male']

for feature in features:
    titanic[feature]= le.fit_transform(titanic[feature])  #객체.fit_transform :인코딩해주는거 (fit:준비-tr:실행)
titanic.head()

```

```python
titanic.embarked.value_counts()
```

```python
titanic.who.value_counts()
```

```python
titanic.adult_male.value_counts()
```

```python
# Q. df_boston 전체 컬럼에 대해 정규화를 수행하세요.
import pandas as pd
from sklearn.datasets import load_boston  

boston= load_boston()
boston.target
# print(boston.DESCR)

df_boston= pd.DataFrame(boston.data, columns=boston.feature_names)
df_boston['TARGET']= boston.target   
df_boston.head()
```

```python
from sklearn.preprocessing import MinMaxScaler 

scaler= MinMaxScaler()   #최소0최대1 정규화

df_boston[:]= scaler.fit_transform(df_boston[:])
df_boston.head()
```

```python
df_boston.mean()
```

```python
# Q. df_boston 전체 컬럼에 대해 표준화를 수행하세요.
# 평균0 표준편차1 (표준전개분포) 로 스케일링 

from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
df_boston[:]= scaler.fit_transform(df_boston[:])
df_boston.head()

```

```python
df_boston.mean()
```
