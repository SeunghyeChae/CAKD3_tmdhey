```python
# 컬럼 추출 , 중복 없애기(unique도 됨/...) , 컬럼값 value 추출 
q=pol[['정체명']]
w=q.drop_duplicates(['정체명'], ignore_index = True)
w.values
```

```python
# 단원제 들어가있으면 값을 단원제로 바꾸고 양원제 들어가있으면 값을 양원제로 바꾸기
import numpy as np
pol['의회형태'] = np.where(pol['의회형태'].str.contains("단원제") == True, "단원제", pol['의회형태'])
pol['의회형태'] = np.where(pol['의회형태'].str.contains("양원제") == True, "양원제", pol['의회형태'])
pol.head()
```

```python
# 정체명- 공화제들어가면 공화제, 공화국-공화국
import numpy as np
pol['정체명'] = np.where(pol['정체명'].str.contains("공화제") == True, "공화제", pol['정체명'])
pol['정체명'] = np.where(pol['정체명'].str.contains("공화국") == True, "공화국", pol['정체명'])
pol.head()
```
