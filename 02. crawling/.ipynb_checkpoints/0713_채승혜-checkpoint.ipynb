```python
# @. 과제  urllib + bs4로 데이터자격검증-빅분기 정보 뽑기

from urllib import request as req
from bs4 import BeautifulSoup

res = req.urlopen("https://www.dataq.or.kr/www/sub/a_07.do")
bs = BeautifulSoup(res,'html.parser')
eles= bs.find_all('div',{'class':['tab-content border-right border-bottom border-left']}) #클래스가 여러개일 경우
for e in eles:
    print(e.text)
```

```python

import time
from bs4 import BeautifulSoup
from selenium import webdriver

url = 'https://play.google.com/store/movies/top'
driver = webdriver.Chrome('C:/♥workspace/tool/chromedriver.exe')
driver.get(url)
time.sleep(2)

#time_sleep 
interval = 2 

# 현재 문서 높이를 가져와서 저장
prev_height = driver.execute_script('return document.body.scrollHeight')

# 반복 수행
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    # 페이지 로딩 대기
    time.sleep(interval)
    curr_height = driver.execute_script('return document.body.scrollHeight')
    if curr_height == prev_height:
        break
    else:
        prev_height = curr_height


bs = BeautifulSoup(driver.page_source, 'html.parser')

#할인되는 영화의 할인 전 가격들 ... 
before=driver.find_element_by_class_name('SUZt4c djCuy')

# after= driver.find_element_by_class_nam('//*[@id="fcxH9b"]/div[4]/c-wiz[3]/div/c-wiz/div/div/c-wiz/c-wiz/c-wiz/div/div[2]/c-wiz/div/div/div[2]/div/div/div[2]/div/div[2]/div/button/div/span[2]/span')
# title_name= driver.find_element_by_class_nam('WsMG1c nnK0zc')


for i in before:
    title_name=before.find('div', attrs = {'class':'WsMG1c nnK0zc'})
    after=before.find('span', attrs = {'class':'VfPpfd ZdBevf i5DZme'})
    link=before.find('a', attrs = {'class':'JC71ub'})
    print ('제목 :{}\n할인 전 금액 :{}\n할인 후 금액 :{}\n링크 {link_t[]}'.format(title_name.text,before.text,after.text,link.text))
    print('\n')
    time.sleep(1)

driver.quit()
```
