```python
# 창 띄우기 
import sys
from PyQt5.QtWidgets import QApplication, QWidget

class MyApp(QWidget):

  def __init__(self):   # self = MyApp 객체
    super().__init__()
    self.initUI()

  def initUI(self):
    self.setWindowTitle('My First Application')  # 타이틀바에 나타나는 창의 제목 설정
    self.move(300,300)  # 위젯을 스크린의 x 300 y 300px 의 위치로 이동
    self.resize(400,200)  # 위젯의 크기를 너비 400 높이 200px로 조절 
    self.show()


if __name__ == '__main__':  # 현재 모듈의 이름이 저장되는 내장 변수 
  app= QApplication(sys.argv) # 어플리케이션 객체 생성 
  ex= MyApp()
  sys.exit(app.exec_())
```

```python
# 어플리케이션 아이콘 넣기
import sys
from PyQt5.QtWidgets import QApplication, QWidget
from PyQt5.QtGui import QIcon

class MyApp(QWidget):

  def __init__(self):   # self = MyApp 객체
    super().__init__()
    self.initUI()

  def initUI(self):
    self.setWindowTitle('Icon')  # 타이틀바에 나타나는 창의 제목 설정
    self.setWindowIcon(QIcon('heart.png')) # 아이콘 설정
    self.setGeometry(300,300,300,200)
    self.show()


if __name__ == '__main__':  # 현재 모듈의 이름이 저장되는 내장 변수 
  app= QApplication(sys.argv) # 어플리케이션 객체 생성 
  ex= MyApp()
  sys.exit(app.exec_())
```

```python
# 창닫기
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton
from PyQt5.QtGui import QIcon
from PyQt5.QtCore import QCoreApplication
from PyQt5 import uic


# UI 연결
form_class = uic.loadUiType("dashboard_pyqt.ui")[0]

#화면을 띄우는데 사용되는 Class 선언
class WindowClass(QMainWindow, form_class) :
    def __init__(self) :
        super().__init__()
        self.setupUi(self)

    def initUI(self):
        btn= QPushButton('Quit',self) # 푸시버튼 만들기
        btn.move(50,50)
        btn.resize(btn.sizeHint())
        btn.clicked.connect(QCoreApplication.instance().quit) # click시그널 만들어짐 . /instance는 현재 인스턴스 반환
                                                            #  clicked 시그널은 어플리케이션을 종료하는 quit메서드에 연결

    
        self.setWindowTitle('Quit')  # 타이틀바에 나타나는 창의 제목 설정
        self.setWindowIcon(QIcon('heart.png')) # 아이콘 설정
        self.show()


if __name__ == '__main__':  # 현재 모듈의 이름이 저장되는 내장 변수 
  app= QApplication(sys.argv) # 어플리케이션 객체 생성 
  ex= MyApp()
  sys.exit(app.exec_())
```

```python
# Qt Designer의 UI는 저장버튼(Control + S / Command + S)를 누르면 저장을 할 수 있습니다.
# 저장된 UI파일은 XML의 형식을 가지며, Python 코드에서 이 XML 파일을 Import한 후 위젯들에 기능을 할당해주면 
# 실제로 기능을 가지고 작동하는 GUI프로그램이 되는 것입니다. 직접 XML형식의 ui파일을 수정하여 
# 레이아웃을 수정할 수도 있습니다.
```

```python
# qt designer 연결하고 quit 버튼 만들기 


import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton
from PyQt5.QtGui import QIcon
from PyQt5.QtCore import QCoreApplication
from PyQt5 import uic
from PyQt5.QtWidgets import *

form_class = uic.loadUiType("dashboard_pyqt.ui")[0]

class WindowClass(QMainWindow, form_class) :
    def __init__(self) :
        super().__init__()
        self.setupUi(self)

        self.setWindowIcon(QIcon('heart.png')) # 아이콘 설정
        
        self.btn_1.clicked.connect(self.button1Function)
        
        btn= QPushButton('Quit',self) # 푸시버튼 만들기
        btn.move(50,50)
        btn.resize(btn.sizeHint())
        btn.clicked.connect(QCoreApplication.instance().quit) # click시그널 만들어짐 . /instance는 현재 인스턴스 반환
                                                            #  clicked 시그널은 어플리케이션을 종료하는 quit메서드에 연결
        
        btn = QPushButton('Button', self)
        btn.setToolTip('This is a <b>QPushButton</b> widget')
        btn.move(50, 50)
        btn.resize(btn.sizeHint())
        btn.clicked.connect(QCoreApplication.instance().button1Function)
        
        #버튼에 기능을 연결하는 코드
        self.btn_1.clicked.connect(self.button1Function)
        self.btn_2.clicked.connect(self.button2Function)

        btn_1이 눌리면 작동할 함수
        def button1Function(self) :
            print("btn_1 Clicked")
        
        
    
        self.setWindowTitle('HIHIHI')  # 타이틀바에 나타나는 창의 제목 설정
        self.setWindowIcon(QIcon('heart.png')) # 아이콘 설정
        self.show()


        
if __name__ == "__main__" :
    app = QApplication(sys.argv)
    myWindow = WindowClass() 
    myWindow.show()
    app.exec_()
```

```python
import sys
from PyQt5 import QtCore, QtGui, QtWidgets, uic
from PyQt5.QtWidgets import QMainWindow, QLabel, QPushButton, \
    QApplication, QFileDialog, QMessageBox
from PyQt5.QtGui import QPixmap

form_class = uic.loadUiType("dashboard_pyqt.ui")[0]


class MainWindow(QMainWindow, form_class):
    def __init__(self):
        super().__init__()
        self.setupUi(self) # ui 설정
        self.setWindowTitle('HIHIHIHI') #타이틀 

        self.img_button.clicked.connect(self.filedialog_open) # 버튼누르면 파일경로 dialog 생성


        # 이미지를 추가할 라벨
        self.image_frame = QLabel(self)

    def filedialog_open(self):
        fname = QFileDialog.getOpenFileName(self, 'Open File', '',
                                            'All File(*);; Image File(*.png *.jpg)')

        if fname[0]:
            # QPixmap 객체
            pixmap = QPixmap(fname[0])
            pixmap = pixmap.scaled(481,621) # 이미지 스케일 변화
            self.image_frame.setPixmap(pixmap)  # 이미지 세팅
           
            self.image_frame.resize(481,621)  # 프레임 스케일
            
#             이미지의 크기에 맞게 Resize
#             self.resize(481,621)    # 전체화면 스케일

        else:
            QMessageBox.about(self, 'Warning', '파일을 선택하지 않았습니다.')


if __name__ == '__main__':
    app = QApplication(sys.argv)
    mainWindow = MainWindow()
    mainWindow.show()
    sys.exit(app.exec_())
```
