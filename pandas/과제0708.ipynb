```python
# 과제
# Q. 5행 5열의 리스트, 배열, 사전 형태의 데이터를 데이터 프레임으로
# 변환하여 출력하세요(가능한한 의미있는 데이터로 작성)


# list
import pandas as pd

score=[(88,97,28,77,66),(100,97,68,68,90),(77,79,68,96,91),(55,60,38,22,70),(100,80,26,74,48)]

print(score, type(score))


sdf = pd.DataFrame(score)
sdf
```

```python
#array
import numpy as np

np.random.seed(0)
score_ = np.random.randint(0,100, size=(5,5))

print(score_, type(score_))

df_1 = pd.DataFrame(score_)
df_1= pd.DataFrame(score_,index=['승해','동연','도연','도환','지환'],
                columns=['국어','수학','영어','체육','과학'])

df_1
```

```python
# dict
height = {0:[158,166,190,180,172],
1:[182,171,168,160,159],
2:[156,175,185,167,150],
3:[179,164,163,155,170],
4:[170,163,132,157,161]}
print(height, type(height))

df_h = pd.DataFrame(height)
df_h
```

```python
# Q. 위에서 변환한 데이터 프레임에서 데이터 변경, 삭제, 전치, 인덱스 등 / 처리를 한 후 다시 리스트, 배열, 사전 형태로 변환하여 출력하세요.

df_hc = df_h.copy()
df_hc
df_hc = df_hc.rename(index={0:'1번',1:'5번',2:'10번',3:'15번',4:'20번'}
                     ,columns={0:'1반',1:'2반', 2:'3반',3:'4반',4:'5반'})
                                                                            
df_hc
```

```python
# 데이터 변경
df_hc.loc['1번']= '*비공개'
df_hc.loc['10번','2반']='**'

df_hc
```

```python
# 데이터 삭제 (행)
df_hc = df_hc.drop(['20번'], axis=0)
df_hc
```

```python
#데이터 삭제 (열)
df_hc= df_hc.drop('2반',axis=1)
df_hc
```

```python
# 전치
df_hc = df_hc.transpose()
df_hc
```

```python
# 인덱스 처리

df_hc.index.name='^~^'
df_hc
```
