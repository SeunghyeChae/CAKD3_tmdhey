```python
import random
import numpy as np
import pandas as pd

np.random.seed(0)

A = np.random.randint(45,100, size=(5,5))
B = np.random.randint(45,100, size=(5,5))

A_df = pd.DataFrame(A)
B_df = pd.DataFrame(B)

print(A_df,type(A_df))
print()
print(B_df,type(B_df))
```

```python
r1 = A_df.iloc[0]
r2 = B_df.iloc[0]
c1 = A_df[0]
c2 = B_df[0]

print(r1)
print()
print(r2)
print()
print(c1)
print()
print(c2)
```
