```python
import random
hanguls = list('가나다라마바사아자차카타파하')
with open('info.txt','w') as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        height = int(random.randrange(140,200))
        weight = int(random.randrange(40,100))
        BMI= round(weight/((height*0.01)**2),2)
        
        if BMI<18.5:
            result='저체중'
        elif 18.5<= BMI <25:
            result='정상'
        elif 25<= BMI <30:
            result='과체중'
        else:
            result='비만'
        
        file.write("{},{},{},{},{}\n".format(name,height,weight,BMI,result))
        
#--------------------------------------------------        
with open('info.txt','r')as file:
    for A in file:
        (name,height, weight,BMI,result)= A.strip().split(',')
        print("\n".join([
            "이름:{}",
            "키:{}",
            "몸무게:{}",
            "BMI:{}",
            "결과:{}"
            ]).format(name,height,weight,BMI,result))
        print()
```
