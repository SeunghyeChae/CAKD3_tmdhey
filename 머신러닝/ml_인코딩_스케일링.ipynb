```python
# 레이블 인코딩 
from sklearn.preprocessing import LabelEncoder

items= ['TV','냉장고','전자레인지','컴퓨터','선풍기','선풍기','믹서','믹서']
encoder = LabelEncoder()
labels= encoder.fit_transform(items)
print('인코딩 변환값:', labels)
```

```python
print('인코딩 클래스:',encoder.classes_)
```

```python
print('디코딩 원본 값:',encoder.inverse_transform([0,1,4,5,3,3,2,2]))
```

```python
# 원-핫 인코딩  
# 숫자의 크기를 해결할 수 있음. but 컬럼이 부풀려짐 (변수가 많아짐), 정확도가 높아짐 (분류모델,회귀모델의 경우)
# 회귀에서는 규제를 가하는 것을 적용 (l2 적용: 릿지모델 //  l1 적용:랏지모델)
# --> 규제로 부작용 보정함 

# 레이블 인코딩 먼저 해줌
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items= ['TV','냉장고','전자레인지','컴퓨터','선풍기','선풍기','믹서','믹서']

encoder= LabelEncoder()
labels= encoder.fit_transform(items)
labels= labels.reshape(-1,1)
labels

oh_encoder= OneHotEncoder()
oh_labels= oh_encoder.fit_transform(labels)
print('원-핫 인코딩:')
print(oh_labels.toarray())
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris
```

```python
iris_df= pd.DataFrame(data= iris.data , columns=iris.feature_names)
iris_df.head()

print('mean:\n',iris_df.mean())
print()
print('std:\n',iris_df.std())
```

```python
# 정규화
from sklearn.preprocessing import MinMaxScaler

scaler_MM= MinMaxScaler()
iris_scale_MM = scaler_MM.fit_transform(iris_df)
iris_scale_MM_df = pd.DataFrame(iris_scale_MM, columns=iris_df.columns)
iris_scale_MM_df
```

```python
print('mean:\n',iris_scale_MM_df.mean())
print()
print('std:\n',iris_scale_MM_df.std())
```

```python
# 표준화 
from sklearn.preprocessing import StandardScaler

scaler_SS = StandardScaler()
iris_scale_SS = scaler_SS.fit_transform(iris_df)
iris_scale_SS_df = pd.DataFrame(iris_scale_SS, columns=iris_df.columns)
iris_scale_SS_df
```

```python
print('mean:\n',iris_scale_SS_df.mean())
print()
print('std:\n',iris_scale_SS_df.std())
```
