## 통계 과제

```python
# 과제1 
# Q. 공분산을 풀어서 구하세요 cov_data 1---- 의미학습 !! 
```

```python
import pandas as pd
import numpy as np

cov_data= pd.read_csv('dataset/4_cov.csv')
cov_data
```

```python
# 1
N= len(cov_data)-1
sum_value= np.sum(cov_data.x)
mu_x= sum_value/N
mu_x
```

```python
sum_value= np.sum(cov_data.y)
mu_y= sum_value/N
mu_y
```

```python
covariance= np.sum(((cov_data.x-mu_x)*(cov_data.y-mu_y))/(N-1))
covariance.round(2)
```

```python
# 2 (간단하게)

N= len(cov_data)
mu_x= cov_data.x.mean()
mu_y= cov_data.y.mean()

covariance=  np.sum(((cov_data.x-mu_x)*(cov_data.y-mu_y))/(N-1))
covariance.round(2)
```

```python
# 과제2 
# Q. 상관계수를 풀어서 구하세요

N= len(cov_data)-1
mu_x= cov_data.x.mean()
mu_y= cov_data.y.mean()
```

```python
X_cov= np.sum((cov_data.x-mu_x)**2/(N-1))
Y_cov= np.sum((cov_data.y-mu_y)**2/(N-1))

rout_XY= (X_cov * Y_cov)**(1/2)

r= covariance/rout_XY
r.round(2)
```

## 시각화 복습 

```python
import seaborn as sns
df= sns.load_dataset('titanic')
df.head()
```

```python
df1= df.drop(['class','who','deck','embark_town','adult_male','alive'],axis=1)
df1.head()
```

```python
df1.age.fillna(df1['age'].mean(),inplace=True)
df1.head()
```

```python
df1.embarked.ffill(inplace=True)
df1.head()
```

```python
df1.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder 

le= LabelEncoder()
features= ['sex', 'embarked','alone']

for feature in features:
    df1[feature]= le.fit_transform(df1[feature]) 

df1.head()

# male:1/ female:0
# C:0 / Q:1 / S:2
# alone x: 0 / alone:1
```

```python
def get_age_cat(age):
    cat=''
    if age<15: cat='young'
    elif age<30: cat='teen'
    elif age<45: cat='adult'
    elif age<60: cat='mature'
    else: cat='elder'
    return cat

df1['age_cat']=df1['age'].apply(lambda x: get_age_cat(x))
df1.head()
```
