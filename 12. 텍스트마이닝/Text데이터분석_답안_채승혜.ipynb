```python
from konlpy.tag import Okt
okt= Okt()
text = """파싱은 프로그램을 compile하는 과정에서 특정 프로그래밍 언어가 제시하는 문법을 잘 지켜서 작성하였는지 compiler가 검사하는 것이다."""

# 명사 
print('명사 >>\n',okt.nouns(text))
print()

# 형태소
print('형태소 >>\n',okt.morphs(text))
print()

# (형태소,품사)
print('(형태소,품사) >>\n',okt.pos(text,norm=True))
print()


# 형태소/품사
print('(형태소,품사) >>\n',okt.pos(text,norm=True,stem=True,join=True))
print()
```

Q4. fetch_20newsgroups 데이터셋을 가져와서 아래와 같이 분류분석을 수행하세요.
- tfidf로 피처 벡터화
- 알고리즘은 LogistiticRegression
- GridSearchCV로 모델 성능 튜닝 'C':[0.01, 0.1, 1, 5, 10]
- 평가는 정확도

```python
from sklearn.datasets import fetch_20newsgroups

news_data = fetch_20newsgroups(subset='all',random_state=156)
```

```python
from sklearn.datasets import fetch_20newsgroups

train_news= fetch_20newsgroups(subset='train',random_state=156)
X_train = train_news.data
y_train = train_news.target

test_news= fetch_20newsgroups(subset='test',random_state=156)
X_test = test_news.data
y_test = test_news.target
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score

tfidf_vect = TfidfVectorizer()
tfidf_vect.fit(X_train)

X_train_tfidf_vect = tfidf_vect.transform(X_train)

X_test_tfidf_vect = tfidf_vect.transform(X_test)

lr = LogisticRegression()

```

```python
params = { 'C':[0.01, 0.1, 1, 5, 10]}
grid_cv_lr = GridSearchCV(lr ,param_grid=params , cv=3 , scoring='accuracy' , verbose=1 )
grid_cv_lr.fit(X_train_tfidf_vect , y_train)
print('최적의 C파라미터:',grid_cv_lr.best_params_ )
print()

pred = grid_cv_lr.predict(X_test_tfidf_vect)
print('예측 정확도 {0:.3f}'.format(accuracy_score(y_test ,pred)))
```

Q5. IMDB 영화평 데이터셋을 불러와서 아래와 같이 감성분석을 수행하세요.
- 지도학습 : 
  - 스톱 워드는 English, ngram은 (1,2)로 설정, CountVectorization수행
  - LogisticRegression의 C는 10으로 설정.
  - Pipeline 객체를 이용하여 fit(), predict()로 학습/예측 수행.
  - 평가는 정확도, ROC 
- 비지도 학습(Vader)
  - Confusion_matrix, 정확도, 정밀도, 재현율로 평가

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score

review_df = pd.read_csv('dataset/labeledTrainData.tsv',header=0, sep="\t", quoting=3)
```

```python
review_df.head(3)
```

```python
review_df.review = review_df.review.str.replace('<br />',' ')
```

```python
def resub(x):
    text = re.sub('[0-9]+','',x)
    text = re.sub('[A-Za-z]+','',text)
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', text)
    return text
```

```python
review_df.review = review_df['review'].apply( lambda x : resub(x) )
```

```python
from sklearn.model_selection import train_test_split

s_df= review_df.sentiment
f_df = review_df.drop(['id','sentiment'], axis=1, inplace=False)

X_train, X_test, y_train, y_test= train_test_split(f_df, s_df,
                                                   test_size=0.3, random_state=156)
```

```python
pipeline = Pipeline([
    ('cnt_vect', CountVectorizer(stop_words='english', ngram_range=(1,2) )),
    ('lr_clf', LogisticRegression(C=10))
])

pipeline.fit(X_train.review, y_train)
pred = pipeline.predict(X_test.review)
pred_probs = pipeline.predict_proba(X_test.review)[:,1]

print('예측 정확도>> {0:.4f} \nROC-AUC>> {1:.4f}'.format(accuracy_score(y_test ,pred),roc_auc_score(y_test, pred_probs)))
```
