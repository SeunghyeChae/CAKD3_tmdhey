```python
#Q.입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성한 후 
# 1,2,3을 입력한 결과를 출력하세요

def QWER(*args):
    result = 0
    for i in args:
        result += i
    return result/len(args)

QWER(1,2,3)
```

```python
# Q. strings를 문자의 크기 순서로 내림차순 정렬

strings= ['a','ca','bar','aaaa','ababa']
strings.sort(key=len,reverse=True)
strings
```

```python
# Q. strings를 다양한 문자가 포함된 순서로 정렬하세요

strings = ['foo', 'card','bar','aaaa','ababa']

strings.sort(key=lambda x : len(set(x)),reverse=True)
print (strings)
```

```python
def getTotalpage(m,n):
    if m%n==0:
        return m//n
    else :
        return m//n +1
    
print(getTotalpage(100,3))
```

```python
import random
hanguls = list('가나다라마바사아자차카타파하')
with open('info.txt','w') as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        height = random.randrange(140,200)
        weight = random.randrange(40,100)
        file.write("{},{},{}\n".format(name,height,weight))
        
with open("info.txt",'r') as file:
    data = file.read()
    print(data)
```

```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns

tt= sns.load_dataset('titanic')
tt.head()
```

```python
def age_cat(age):
    cat=''
    if age<=5: cat='baby'
    elif age<=12: cat='child'
    elif age<=18: cat='Teenager'
    elif age<=35: cat='Young Adult'
    elif age<=60: cat='Adult'
    else:cat='Elderly'
    return cat

tt['age_cat']=tt['age'].apply(lambda x:age_cat(x))
tt.head()
```

```python
tt.age_cat.unique()
```
